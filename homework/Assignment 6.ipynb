```python
import numpy as np
import pandas as pd
```

```python
#loading my environmental data
###set the first line as col names
banner4=pd.read_csv('../AStahl Dataset/PhysicalData_9.23.2019 .csv', nrows=5, encoding='latin')
colnames4 = list(banner4.columns)
print(colnames4)
###read in dataset with the col names that was just made
environ_data=pd.read_csv('../AStahl Dataset/PhysicalData_9.23.2019 .csv', encoding='latin', 
                         names=colnames4, skiprows=2)
###remove the last unwated 'notes' rows/unwanted data
environ_data=environ_data.iloc[:,:-7]
###only pulling the columns I want to use 
environ_data=environ_data.iloc[:,[3,6,7]]
environ_data.head()
####GOOD TO GO
```

```python
#loading my chlorophyll data 
###set the first line as col names
banner5=pd.read_csv('../AStahl Dataset/chldata_9.23.2019.csv')
colnames5 = list(banner5.columns)
print(colnames5)
###read in dataset with the col names that was just made
chl_data=pd.read_csv('../AStahl Dataset/chldata_9.23.2019.csv', 
                     names=colnames5,skiprows=3)
###remove the last unwated 'notes' rows
chl_data=chl_data.iloc[:,:-17]
###only pulling the columns I want to use
chl_data=chl_data.iloc[:,[12,13,14,15,16]]
##Rename columns
chl_data.rename(columns={'Sample Date.1':'Date','surface chla all':'Surface Chla all',
                        'Surface Phaeo all.1':'Surface Phaeo all','Bottom Chla all.1':'Bottom Chla all',
                         'Bottom Phaeo all.1':'Bottom Phaeo all'}, inplace=True)
chl_data.head()
###GOOD TO GO
```

```python
#loading my silica data
###set the first line as col names
banner6=pd.read_excel('../AStahl Dataset/nutrientdata.xls',sheet_name=1)
colnames6=list(banner6.columns)
print(colnames6)
###read in dataset with the col names that was just made
Si_data=pd.read_excel('../AStahl Dataset/nutrientdata.xls',sheet_name=1, names=colnames6,
                     skiprows=2)
#rename columns
Si_data.rename(columns={'DATE':'Date','Si':'Surface Si','Si.1':'Bottom Si'}, inplace=True)
##remove unwanted columns by keeping the columns you want 
Si_data=Si_data.iloc[:,[1,3,9]]
Si_data.head()
###GOOD TO GO
```

```python
#loading my nitrogen and phosphorus data
###set the fifth line as col names
banner7=pd.read_excel('../AStahl Dataset/nutrientdata.xls',sheet_name=0, header=4)
colnames7=list(banner7.columns)
print(colnames7)
###read in dataset with the col names that was just made
NandP_data=pd.read_excel('../AStahl Dataset/nutrientdata.xls',sheet_name=0, names=colnames7,
                         skiprows=6)
###rename columns
NandP_data.rename(columns={'Unnamed: 0':'Date','NH4':'Surface NH4','DIP':'Surface DIP',
                           'NO3+2':'Surface NO3+2','NO3':'Surface NO3','NO2':'Surface NO2',
                          'DIN':'Surface DIN','NH4.1':'Bottom NH4','DIP.1':'Bottom DIP',
                           'NO3+2.1':'Bottom NO3+2','NO3.1':'Bottom NO3','NO2.1':'Bottom NO2',
                           'DIN.1':'Bottom DIN'}, inplace=True)
##remove unwanted columns by keeping the columns you want 
NandP_data=NandP_data.iloc[:,[0,2,3,4,5,6,7,11,12,13,14,15,16]]
NandP_data.head()
##GOOD TO GO
```

```python
#loading my count data
###set the first line as col names
banner8=pd.read_excel('../AStahl Dataset/countdata_9.23.2019 .xls',sheet_name=1)
colnames8=list(banner8.columns)
###read in dataset with the col names that was just made
count_data=pd.read_excel('../AStahl Dataset/countdata_9.23.2019 .xls',sheet_name=1, 
                         names=colnames8, skiprows=1)
#rename columns
count_data.rename(columns={'DATE':'Date'}, inplace=True)
##remove unwanted columns by keeping the columns you want 
count_data=count_data.iloc[:,[0,3,8,214,215,216,217,218,234,245,248]]
count_data.head()
```

```python
#merging some above datasets
data1=pd.merge(environ_data, chl_data, on=['Date'])
data1.head()
#fixing the Date format for environ and chl
data1['Date'] =  pd.to_datetime(data1['Date'],
                              format='%d-%b-%y')
data1
print(data1['Date'])
```

```python
#merging more datasets
data2=pd.merge(Si_data, NandP_data, on=['Date'])
data2.head(15)

#fixing the date format for Si and N/P
data2['Date'] =  pd.to_datetime(data2['Date'],
                              format='%Y-%m-%d')
print(data2['Date'])
```

```python
#merging the first 4 datasets
data3=pd.merge(data1,data2, on=['Date'])
data3
```

```python
#fixing the count data date format
count_data['Date']= pd.to_datetime(count_data['Date'],errors='coerce')
count_data.info()
count_data.head()
```

```python
#final merge of all datasets
data=pd.merge(data3,count_data, on=['Date'])
data.head()
```

```python
#I am not looking to fill in any data so I don't want to use the  fillna function
#If I am looking to compare changes over time, I need all the values for all of the variables
#therefore I think the best way to be able to collect all that data is to remove the entire row
data.dropna(how='all')
# I do not think that the data.dropna() will give me enough resolution for the test I want to do 
#Can I do statistical analysis with the data if there is NaN in it? Or does NaN HAVE to be removed 
     ##to do statistical analysis?
```

The null hypothesis for my questions is that there will be no change in temperature/environmental factors or cell abundance over time. 
A t-test is used to determine if there is a statistical difference between the means of 2 groups. 
A p-value is the set level used to determine if there is a significant statistical result from your analysis (usually 0.05). The p-value is the main source of information for accepting or rejecting your null hypothesis. 
The standard deviation tells you how far your data spreads out around the mean. To calculate it:
     find the mean of your dataset
     subtract the mean from each data point 
     square the result
     find the mean of the squared results (this number is known as the variance)
     take the square root of the variance and that is your SD.
