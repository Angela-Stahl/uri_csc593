```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
data=pd.read_csv('../AStahl Dataset/all_data.csv')
data['Date'] =pd.to_datetime(data['Date'], format='%Y-%m-%d')
data.info()
```

```python
#set the size
fig, ax1 = plt.subplots(figsize=(15,8))

#setting the x axis label for both y axes
ax1.set_xlabel('Date')
ax2=ax1.twinx()

#plotting the temperature on left y axis
ax1.plot(data['Date'].dt.year, data['Surface Temp'], color='r')
#setting the left y axis label
ax1.set_ylabel('Surface Temperature (°C)', c='r',size='large')
#changing the left y axis tick marks to red
ax1.tick_params(axis='y', labelcolor='r')

#plotting the total abundance on right y axis
ax2.plot(data['Date'].dt.year, data['Total abundance'], color='g')
#setting the right y axis label
ax2.set_ylabel('Total Phytoplankton Abundance (cells/L)', c='g',size='large')
#changing the left y axis tick marks to green
ax2.tick_params(axis='y', labelcolor='g')

#changing title
plt.title('Change in Temperature and Chlorophyll a Over Time',size='x-large')

#changing frequency of ticks on x axis
ax1.set_xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018])

plt.show()
#Too crowded for me
```

```python
#Same as above graphs just plotted on 2 different graphs
#set the size and subplots
fig, axs = plt.subplots(1, 2, figsize=(15,8))

#plotting data, setting titles/axes
axs[0].plot('Date', 'Surface Temp', 'r', data=data)
axs[0].set_title('Change in Surface \n Temperature Over Time', loc="center",size='large')
axs[0].set_ylabel('Surface Temperature (°C)',size='medium')
axs[0].set_xlabel('Date')

axs[1].plot('Date','Total abundance' ,'g', data=data)
axs[1].set_title('Change in Total Phytoplankton \n Abundundance over Time', loc="center",size='large')
axs[1].set_ylabel('Total Phytoplankton Abundance (cells/L)',size='medium')
axs[1].set_xlabel('Date')
```
